## Básico da linguagem

Abaixo o exemplo mais simples de um programa em python

```python
print ("Alo mundo!")
```

Adicionalmente, variáveis podem ser utilizadas para armazenar valores em seu programa.

```python
txt = "Boa copa do mundo"
print (txt)

ano = 2014
ano = ano + 4
print(2018)
type(ano)
```

Observer que o tipo associado a estas variáveis é dinâmico em Python.

Além de tipos primitivos, temos também algumas estruturas de dados na linguagem, como lista, tupla e dicionarios. Cada tipo de dado possui um conjunto de operações associado a ele.

Abaixo, um exemplo de uma lista com sequência dos anos das últimas copas do mundo.


```python
anosCopaMundo = [1988,1992,1996,2010,2014]
print ('lista antes: ', anosCopaMundo)
anosCopaMundo.append(2018)
print('lista depois: ', anosCopaMundo)
print( anosCopaMundo[1], ' - ', anosCopaMundo[5])
```

A qualquer hora, você pode obter informações sobre uma função através do comando help.

```python
help(anosCopaMundo.append)
```

O comando append modifica a lista, adicionando um novo elemento ao final da sequência.
Abaixo um exemplo de um comando que não modifica a lista.

```python
anosCopaMundo.append(1984)
print ('lista após a adição de 1984 ', anosCopaMundo)
sorted(anosCopaMundo)
print('lista inalterada ', anosCopaMundo)
anosCopaMundo = sorted(anosCopaMundo)
print('lista ordenada ', anosCopaMundo)
```

Abaixo, um exemplo de tupla com informações (de diferentes tipos) sobre o apresentador.

```python
tuplaPessoa = ('Rodrigo', 38, 1.69)
print (tuplaPessoa)
print (tuplaPessoa[0])
```

Abaixo, um exemplo de dicionário com as informações anteriores de maneira mais estruturada.

```python
dadosPessoa = {'nome': "Rodrigo", 'idade': 38, 'altura': 1.69}
print (dadosPessoa['nome'])
print('No próximo ano ', dadosPessoa['nome'], ' fará ', dadosPessoa['idade']+1 , ' anos')
```

Loops em uma estrutura de dados são feitos de maneira simples, como mostrado abaixo.
Observe que em python o comando for é acompanhando sempre de ':' para indicar as instruções
inclusas detro do loop. Tais instruções devem ser obedecer um mesmo recuo (identação).

```python
ingredientesBolo = ['farinha', 'fermento', 'leite' , 'ovos', 'açucar', 'manteiga']
print('Os ingredientes de um bolo são: ')
for item in ingredientesBolo:
    print(item)
```

Por simplicidade, Python permite a construção de listas a partir da iteração de outras listas.
No exemplo abaixo, a lista anosAposCopa é definida pelos anos de cada uma das copas do mundo mais 1.

```python
anosAposCopa = [ ano+1 for ano in anosCopaMundo]
print (anosAposCopa)
```

Para definir uma função utilize a palavra chave 'def'.
Como visto no último exemplo, o bloco de instruções de uma função possuem um recuo de epaços ("tab")

```python
def potQuadrado(x):
    potencia = x*x
    return potencia
potQuadrado(3)
```

```python
def calc(x, y, op):        
    if op == 'soma':         
        return x + y
    elif op == 'subtracao':
        return x - y
    else:
        print ('Operação inválida')
calc(5,3, 'multiplicacao')
calc(5,3, 'soma')
calc(5,3, 'subtracao')
```

#### Pequeno exemplo que faz o download da lista de deputados do estado

Após vermos alguns exemplos básicos da linguagem. Mostramos abaixo um pequeno programa que recupera a lista dos deputados federais atuais (disponível no site da Câmera Federal) do estado do Rio Grande do Norte.
Obs.: Neste exemplo, mostramos também um exemplo de importação e uso de modulos.

```python
import requests     # import module para fazer o download de conteúdo na internet
import pprint
r = requests.get('https://dadosabertos.camara.leg.br/api/v2/deputados?siglaUf=RN&itens=550&ordenarPor=nome')
top = r.json()      # converte um conteúdo JSON e o armazena em um dicionário
# pprint.pprint (top)  # descomente esta linha para ver o conteúdo baixado do site da camera
nomesDeputados = [deputado['nome'] for deputado in top['dados']]
nomesDeputados = sorted (nomesDeputados)
print(nomesDeputados)
```

## Introdução a pandas

Nesta parte do workshop, utilizaremos um exemplo apresentado por Kevin Markham em https://github.com/justmarkham/python-data-analysis-workshop
    
No exemplo, apresentamos os dados de consumo de bebidas alcoólicas (cerveja, vinho e destilados) nos diversos paises. Estes dados foram extraidos de:
http://fivethirtyeight.com/datalab/dear-mona-followup-where-do-people-drink-the-most-beer-wine-and-spirits/

```python
# importação das bibliotecas necessárias
import pandas as pd                 # principal biblioteca para análise de dados dados
import numpy as np                  # biblioteca para computação ciêntífica
import matplotlib.pyplot as plt     # geração e apresentação de gráficos
```

Abaixo, realizamos o download de uma planilha com os mesmo dados do artigo da fivethirtyeight

```python
drinks = pd.read_csv('https://raw.githubusercontent.com/justmarkham/python-data-analysis-workshop/master/drinks.csv')
type(drinks)
```

Observe que o tipo de dado retornado é próprio da bibliteca pandas.
Para examinar rapidamente os dados, basta que executemos os comandos abaixo. 

```python
#drinks                 # imprime as primeiras e as últimas 30 linhas
drinks.head()           # imprime as primeiras 5 linhas
#drinks.tail()           # imprime as últimas 5 linhas
#drinks.info()          # sumariza os dados
#drinks.describe()       # exibe uma rápida descrição estatística dos dados
```

Infelizmente, como em vários casos reais, nem sempre recebemos os dados 'prontos' para uso.
Manipulações e limpezas são frequêntemente necessárias.

```python
drinks.isnull()         # Mostra que células não possuem valores
drinks.isnull().sum()   # realiza o somatório de células sem valores em cada coluna
```

Neste momento, podemos tomar várias decisões sobre como tratar os dados.
Abaixo, algumas das mais simples e ingênuas.

```python
drinks.dropna()             # remove qualquer valor que estiver faltando
drinks.fillna(value='NA')   # preenche os valores ausêntes
```

```python
drinks = drinks.dropna()
drinks.isnull().sum()
```

```python
drinks.continent.value_counts()
```

Vamos adicionar uma nova coluna com o total de todas as bebidas alcoólicas consumidas por cada país.

```python
drinks['total_servings'] = drinks.beer_servings + drinks.spirit_servings + drinks.wine_servings
drinks.head()
```

Abaixo alguns filtros e ordenações que podemos executar.

```python
# os 5 principais paises que consomem mais vinho que cervejo
drinks[drinks.wine_servings > drinks.beer_servings].head()
```

```python
# Os paises que menos consomem cerveja
drinks.sort_values(by='beer_servings').tail()
```

```python
# Média de consumo de cerveja em cada continente
drinks.groupby('continent').beer_servings.mean()
```

#### Geração de gráficos

```python
# exibição do número de paises em cada continente
fig1 = plt.figure()
drinks.continent.value_counts().plot(kind='bar', title='Countries per Continent')
#plt.xlabel('Continent')
#plt.ylabel('Count')
fig1.show()
```

```python
# exibição da média de cerveja consumida em cada continente
fig2 = plt.figure()
drinks.groupby('continent').beer_servings.mean().plot(kind='bar')
fig2.show()
```

```python
# histogram of beer servings
fig3 = plt.figure()
drinks.beer_servings.hist(bins=20)
fig3.show()
```

```python
fig4 = plt.figure()
colors = np.where(drinks.continent=='EU', 'r', 'b')
drinks.plot(x='beer_servings', y='wine_servings', kind='scatter', c=colors)
fig4.show()
```
